```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import gmaps

# Google API Key
from api_keys import g_key
```

```python
########################################################################################
# The file referenced in the first line is too big to be uploaded to github.
# For purposes of checking the code, change the first line of code by removing the #
# and comment out the existing line.
########################################################################################

#bike_trip_data_df = pd.read_csv("trip_data/201912-capitalbikeshare-tripdata.csv")

bike_trip_data_df = pd.read_csv("trip_data/combined_df.csv")

station_number = bike_trip_data_df['Start station number']
station_address = bike_trip_data_df['Start station']

bike_trip_location = pd.DataFrame(station_number)
bike_trip_location["Station Address"] = ""
bike_trip_location["Lat"] = ""
bike_trip_location["Lng"] = ""
bike_trip_location["Station Address"]= station_address

bike_trip_location_unique = bike_trip_location.drop_duplicates()
```

```python
bike_trip_location_unique.head()
```

```python
#find a lat/lon to each station number from google
city = "Washington DC"
params = {"key": g_key}

for index, row in bike_trip_location_unique.iterrows():
    


    target_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
 

    address = row['Station Address']
    params['address'] = address,city
    
    station_lat_lng = requests.get(target_url,params = params)
    
    station_lat_lng = station_lat_lng.json()
    try:
        bike_trip_location_unique.loc[index, "Lat"] = station_lat_lng["results"][0]["geometry"]["location"]["lat"]
        bike_trip_location_unique.loc[index, "Lng"] = station_lat_lng["results"][0]["geometry"]["location"]["lng"]
        
    except:
        print("Station Not Found... Skipping....")
        bike_trip_location_unique.loc[index, "Lat"] = None
        bike_trip_location_unique.loc[index, "Lng"] = None    #station_lat_lng
```

```python
# Merge two dataframes using an inner join
bike_data_merge = pd.merge(bike_trip_location, bike_trip_location_unique, on="Start station number")
bike_data_merge = bike_data_merge.rename(columns={"Lat_y": "Lat", "Lng_y": "Lng"})
bike_data_merge.head()
```

```python
#find the most popular start location(station number)
#start_end_location.set_index(['Start station number']).count()
df_grpby = bike_data_merge.groupby(['Start station number']).agg({'Station Address_x':'count','Lat':'first','Lng':'first'})
#pop_start = start_end_location(["Start station number"]).count()
#start_end_location.head(25)
df =pd.DataFrame(df_grpby)
df =df.rename(columns={'Station Address_x':'rentals'})
df =df.sort_values(by='rentals', ascending=False)
df =df[df['rentals'] > 30000]
df.head(25)
```

```python
df =df.dropna()
```

```python
#heat map

# Store latitude and longitude in locations
locations = df[["Lat", "Lng"]]

# Fill NaN values and convert to float
rentals = df["rentals"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rentals, 
                                 dissipating=True, max_intensity=240000,
                                 point_radius=6, opacity=1)


# Add layer
fig.add_layer(heat_layer)

fig.add_layer(gmaps.transit_layer())

# Display figure
fig
```
